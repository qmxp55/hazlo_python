Esto es una ilustracion de como se ve la siguiente integral:

$\displaystyle{\int}_{\pi/2}^{3\pi/2} \sin(x) + \cos(2x)  \, dx$

```python
f = lambda x: np.sin(x**2) + np.cos(2*x)

x = np.linspace(0, 2*np.pi, 100)
y = f(x)
plt.plot(x,y)
```

```python
X = np.linspace(np.pi/2,3*np.pi/2,100)
Y = f(X)
plt.fill_between(X,Y)
plt.plot(x,y)

plt.xticks([np.pi/2,3*np.pi/2], ['$\pi/2$','$3\pi/2$'])
plt.yticks([])
plt.show()
```
