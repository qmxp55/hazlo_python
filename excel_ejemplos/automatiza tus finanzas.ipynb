```python
import glob
import pandas as pd
import numpy as np

archivos = glob.glob('../data/gastos/*')
archivos
```

```python
#
pinturas = {'meses':[], 'precios':[]}
cols = ['it', 'denominacion', 'unidad', 'cantidad', 'precio', 'total']

for hoja in archivos:
    
    df = pd.read_excel(hoja,
                      sheet_name='Detalle de Presupuesto',
                      skiprows=8,
                      names=cols,
                      usecols=np.arange(1,7),
                      header=None)
    
    mascara = df['denominacion'] == 'Pintura'
    gasto = df[mascara]['total'].iloc[0]
    mes = hoja.split('_')[1].split('.')[0]
    
    pinturas['precios'].append(gasto)
    pinturas['meses'].append(mes)
    
df = pd.DataFrame(pinturas)

df    
```

```python
total = df['precios'].sum()
df['%'] = np.round(100 * df['precios'] / total, 1)
df
```
