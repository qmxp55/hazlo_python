## 1) Carga los archivos de excel y unirlos en una sola tabla

## Crear 1000 tablas dinamicas

```python
mask = np.zeros_like(df3['ID'], dtype=bool)
subs = ['Action', 'Adventure', 'Drama', 'Romance', 'War', 'Mystery']
for i in subs:
    mask |= df3['Subject'] == i
    
mask &= df3['Year'] > 1983

#save to files
for i in range(1000):
    df3[mask].to_excel('../data/pivot_files/sample_%s.xlsx' %(str(i)), index=False)
```

```python
for i in range(1000):
    df = pd.read_excel('../data/pivot_files/sample_%s.xlsx' %(str(i)))
    pivot = df.pivot_table(values='Length', index='Year', 
                           columns='Subject').fillna(0)  
    print('============== Tabla %i ==============' %(i))
    print(pivot)
```
