```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
archivo = '../data/salarios.csv'
cols = ['nombre', 'entrada', 'salida', 'salario/h']
df = pd.read_csv(archivo, usecols=cols)
```

```python
df
```

```python
horas = []
for i in range(len(df['nombre'])):
    sal = datetime.strptime(df['salida'][i], "%H:%M:%S")
    ent = datetime.strptime(df['entrada'][i], "%H:%M:%S")
    diff = (sal - ent)
    horas.append(diff.total_seconds()/3600)
```

```python
df['horas'] = np.round(horas, 1)
df['salario'] = df['horas'] * df['salario/h']
df
```
